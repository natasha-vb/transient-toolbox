```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Sun
plt.plot(0,0,0, marker='*', markersize=10)

# Galactic Centre
plt.plot(0,8.3,0, marker='o', markersize=20)
plt.plot([0,0], [0,8.3], [0,0,], ls='--', c='orange')

# AT2022kak
plt.plot(7.88, 6.15, 3.05, marker='X', markersize=10)
plt.plot([0,7.88], [0,6.15], [0,3.05],ls='--', c='g')

plt.plot([0,7.88], [0,6.15], [0,0], ls='-', c='g')
plt.plot([0,7.88], [8.3,6.15], [0,0], ls='-', c='g')
plt.plot([7.88,7.88], [6.15,6.15], [0,3.05], ls='-', c='g')

ax.set_xlim([-5,12])
ax.set_ylim([-3,10])
ax.set_zlim([0,3])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_box_aspect(None, zoom=0.95)
```
